# توزیع داده های categorical

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
df
```

# countplot()

```python
plt.figure(figsize=(8,6) ,dpi=150)
sns.countplot(data=df , x='division')
```

```python
plt.figure(figsize=(8,6) , dpi=150)
sns.countplot(data=df , x='level of education', hue='training level' )
sns.set()
```

# barplot()

```python
plt.figure(figsize=(8,6) , dpi=150)
sns.barplot(data=df , x= 'level of education' , y='salary' , width=0.4 )
```

# estimator() by mean & errorbar() by confidence interval

```python
plt.figure(figsize=(8,6) , dpi=200)
sns.barplot(data=df , x='level of education' , y='salary' , hue='division' , estimator=np.mean , errorbar=('ci' , 95))
```

# plt.legend()

```python
plt.figure(figsize=(8,6) , dpi=200)
sns.barplot(data=df , x='level of education' , y='salary' , hue='division' )
plt.legend(bbox_to_anchor=(1.05, 0.25))
```

# توزیع داده های دسته ای

## boxplot()

```python
plt.figure(figsize=(11,3) , dpi=150)
sns.boxplot(data=df , x='parental level of education' , y='math score')
sns.set()
```

```python
plt.figure(figsize=(11,3) , dpi=150 )
sns.boxplot(data=df , x='parental level of education' , y='math score' , hue='gender')
plt.legend(bbox_to_anchor=(1.05,0.4) , borderaxespad=0 )
```

```python
plt.figure(figsize=(6,5) , dpi=150)
sns.boxplot(data=df , x='math score' , y='parental level of education' , hue='gender', orient='h' , width=0.4)
```

# violinplot()

```python
plt.figure(figsize=(10,6) , dpi=150)
sns.violinplot(data=df , x='parental level of education' , y='math score')
```

# split vs dodge

```python
plt.figure(figsize=(10,6) , dpi=150)
sns.violinplot(data=df , x='parental level of education' , y='math score' , hue='gender' , split=True)
```

# swarmplot

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=df , x='math score')
```

```python
plt.figure(figsize=(7,5), dpi=150)
sns.swarmplot(data=df , x='race/ethnicity' , y='math score' , hue='gender' , dodge=True)
```
